```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

matplotlib.style.use('ggplot')
```

```python
fac = pd.read_csv('facup_v2.csv')

fac.info()
```

```python
teams_per_year = fac.groupby('year')['team'].nunique().reindex(range(1872,2016))
```

```python
%matplotlib inline

matplotlib.style.use('ggplot')

ax = teams_per_year.plot(kind='bar', figsize=(14,3), width=1)

ticks = ax.xaxis.get_ticklocs()
ticklabels = [l.get_text() for l in ax.xaxis.get_ticklabels()]
ax.xaxis.set_ticks(ticks[::5])
ax.xaxis.set_ticklabels(ticklabels[::5])

plt.show()
```

```python
teams_per_stage = fac.groupby(['year', 'stage'])['team'].nunique().unstack().reindex(range(1872,2016))

teams_per_stage
```

```python
tps = teams_per_stage.plot(kind='bar',
                            stacked = True,
                            figsize=(14,6), 
                            width=1,
                            colormap='viridis')

ticks = tps.xaxis.get_ticklocs()
ticklabels = [l.get_text() for l in tps.xaxis.get_ticklabels()]
tps.xaxis.set_ticks(ticks[::5])
tps.xaxis.set_ticklabels(ticklabels[::5])

plt.show()
```

Graph showing each number of teams in each 'stage' per year, however due to the numbering convention it would actually be preferable if all the finals had a value of 14 and counted down. 

```python
fac1 = fac[['round', 'year', 'stage','team']].copy()
```

```python
def set_stages(grp):
    stages = grp['stage'].max()
    grp['stage_r'] = grp['stage'] - stages + 14
    return grp

fac2 = fac1.groupby(['year']).apply(set_stages)
```

```python
teams_per_stage_r = fac2.groupby(['year', 'stage_r'])['team'].nunique().unstack().reindex(range(1872,2016))
```

```python
tpsr = teams_per_stage_r.plot(kind='bar',
                            stacked = True,
                            figsize=(12,4), 
                            width=1,
                            colormap='viridis')

ticks = tpsr.xaxis.get_ticklocs()
ticklabels = [l.get_text() for l in tpsr.xaxis.get_ticklabels()]
tpsr.xaxis.set_ticks(ticks[::5])
tpsr.xaxis.set_ticklabels(ticklabels[::5])
tpsr.legend().set_visible(False)

plt.show()
```

```python
teams_per_stage_i = teams_per_stage_r * -1
```

```python
teams_per_stage_i.head()
```

```python
tpsi = teams_per_stage_i.plot(kind='bar',
                            stacked = True,
                            figsize=(12,4), 
                            width=1,
                            colormap='viridis')

ticks = tpsi.xaxis.get_ticklocs()
ticklabels = [l.get_text() for l in tpsi.xaxis.get_ticklabels()]
tpsi.xaxis.set_ticks(ticks[::5])
tpsi.xaxis.set_ticklabels(ticklabels[::5])
tpsi.legend().set_visible(False)

plt.show()
```

```python
# set stages in reverse order i.e. all finals have same stage no.

fac_stage = fac[['round', 'year', 'stage','team', 'qualifier']].copy()

def set_stages(grp):
    stages = grp['stage'].max()
    grp['stage_r'] = grp['stage'] - stages + 14
    return grp

fac_stage_r = fac_stage.groupby(['year']).apply(set_stages)

fac_stage_r.head()
```

```python
qualifiers = fac_stage_r[fac.qualifier == True]

qualifiers.head()
```

```python
non_qualifiers = fac_stage_r[fac.qualifier == False]

non_qualifiers.head()
```

```python
q_per_year = (qualifiers.groupby(['year', 'stage_r'])['team'].nunique().unstack().reindex(range(1872,2016)) * -1)

nq_per_year = non_qualifiers.groupby(['year', 'stage_r'])['team'].nunique().unstack().reindex(range(1872,2016))
```

```python
%matplotlib inline

q_per_year.sort_index(axis=1, ascending = False, inplace=True) #reverse order of columns for negative stack correct order

q_per_year.head()
```

```python
q_per_year.plot(kind='bar',
                stacked = True,
                figsize=(12,7), 
                width=1,
                colormap='gray_r')

ax = plt.gca()

nq_per_year.plot(kind='bar',
                stacked = True,
                figsize=(12,7), 
                width=1,
                colormap='Blues_r',
                ax = ax)

ticks = ax.xaxis.get_ticklocs()
ticklabels = [l.get_text() for l in ax.xaxis.get_ticklabels()]
ax.xaxis.set_ticks(ticks[::5])
ax.xaxis.set_ticklabels(ticklabels[::5])
ax.legend().set_visible(False)

plt.show()
```
