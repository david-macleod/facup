```python
yr_1990 = fac[fac.year == 1990]

yr_1990[yr_1990['round'] == 'Semi-Final']['stage'].max()
```

```python
def final_stage(grp):
    grp.ix[(grp['round'] == 'Final'), 'stage'] = grp.ix[(grp['round'] == 'Semi-Final'), 'stage'].max() + 1
    return grp
```

```python
fac_tst = fac.groupby('year').apply(final_stage) #for the stage value of each final add 1 to the semi-final value
```

```python
fac_tst.info()
```

```python
fac_tst['stage'].value_counts()
```

```python
fac['stage'].value_counts()
```

```python
#now want to concatenate a copy of the datafram with itself and create a new 'team' column
#also want to creat a 'home' column (boolean) to easily filter matches
```

```python
fac_home = fac_tst.copy()
fac_away = fac_tst.copy()
```

```python
fac_home['team'] = fac['home_team']
fac_away['team'] = fac['away_team']

fac_home['home'] = True
fac_away['home'] = False
```

```python
fac['home_team'].nunique(), fac['away_team'].nunique()
```

```python
fac_home['team'].nunique(), fac_away['team'].nunique()
```

```python
fac_join = pd.concat([fac_home, fac_away])
```

```python
fac_join['team'].nunique(), fac_join['home_team'].nunique(), fac_join['away_team'].nunique()
```

```python
fac_join.to_csv('facup_v2.csv', index=False)
```
