```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

%matplotlib inline
matplotlib.style.use('ggplot')
```

```python
fac = pd.read_csv('facup_v3.csv')
```

```python
fac.info()
```

```python
fac.head()
```

```python
fac4 = fac[fac.home == False]
```

```python
fac4.info()
```

```python
fac4 = fac4[['home_team','away_team','home_score', 'away_score','round','year','replay','qualifier', 'stage_r','match_id']]
```

```python
fac4.info()
```

```python
fac4.head()
```

```python
fac4['stage'] = 15-fac4.stage_r
```

```python
fac4.drop('stage_r',axis=1, inplace=True)
```

```python
fac4.head()
```

```python
fac4.home_team = fac4['home_team'].str.upper()
fac4.away_team = fac4['away_team'].str.upper()
```

```python
fac4.head()
```

```python
fac4.home_team[fac4.home_team.str.contains(' \(\d\)')].unique()  #looking for (#) with one or more values between brackets
```

```python
fac4.home_team = fac4.home_team.str.replace(' \(\d\)','')
fac4.away_team = fac4.away_team.str.replace(' \(\d\)','')
```

```python
fac4.home_team[fac4.home_team.str.contains(' \(\d\)')].unique()  #looking for (#) with one or more values between brackets
```

```python
fac4.head()
```

```python
fac4['winner'] = np.where(fac4.home_score > fac4.away_score, fac4.home_team, fac4.away_team)
```

```python
fac4.head()
```

```python
fac4['winner'] = np.where(fac4.home_score == fac4.away_score, np.nan, fac4.winner) #testing for draws
```

```python
fac4.head(20)
```

```python
fac4['left'] = fac4[['home_team','away_team']].max(axis=1)
fac4['right'] = fac4[['home_team','away_team']].min(axis=1)
```
