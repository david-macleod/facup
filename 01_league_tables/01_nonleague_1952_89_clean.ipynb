#### Searching for nearest match for each team within the SAME year to try and tie down any duplicate/reserve teams

```python
from fuzzywuzzy import process
import fuzzywuzzy
import Levenshtein
```

```python
tab_1954 = nl_tab[nl_tab.year == 1954].copy()
```

```python
def near_match(row):
    matches =  process.extract(row['team'], tab_1954[tab_1954.year == row['year']]['team'].values, limit = 2) # only search specific year
    # matches is list of tuples in form [('1st nearest match, 1st nearest match score), (2nd nearest match, 2nd nearest match score)]
    return matches[1]
    #return 2nd nearest match 

#%time tab_1954.apply(near_match, axis = 1)
%time tab_1954['match'], tab_1954['score'] = zip(*tab_1954.apply(near_match, axis = 1))
```
