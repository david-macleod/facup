```python
fl = pd.read_csv('football_league_v1.csv', na_values = '')
pl = pd.read_csv('premier_league_v1.csv', na_values = '')
```

```python
df = pd.concat([fl, pl], ignore_index = True) 
```

```python
df2 = df.drop(['Promotion or relegation', 
               'Promotion, qualification or relegation', 
               'Qualification or relegation', 
               'Notes',
               'GD'], axis=1)
```

```python
#normalize: convert non-breaking spaces to spaces, convert to uppercase and strip spaces from ends


df2['team'].replace('\xc2\xa0', ' ', inplace = True, regex = True)
df2['team'] = df2['team'].str.upper().str.strip()
```
