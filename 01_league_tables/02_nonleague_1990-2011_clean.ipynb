```python
import pandas as pd
import numpy as np
```

```python
nl = pd.read_csv(r'C:\Users\David\Python\scrapy\facup\01_league_tables\nonleague_1990-11.csv')
```

```python
nl.info()
```

```python
nl.columns = nl.columns.str.lower()
```

```python
unn_rows = ~nl['unnamed: 0'].isnull()  

nl[unn_rows]
```

For these league tables pos and team need copying into the correct column

```python
nl.ix[unn_rows, 'team'] = nl.ix[unn_rows, 'unnamed: 1']
nl.ix[unn_rows, 'pos'] = nl.ix[unn_rows, 'unnamed: 0']
```

```python
nl.info()
```

Only need to keep year, league, post and team for this project, might clean other columns at a later stage

'team' seems to be missing 110 values, possibly in 'club'

```python
nl[~nl.club.isnull()][['club', 'team']].head()
```

```python
nl[~nl.club.isnull()][['club', 'team']].info()
```

The missing teams correspond with the club column

Filling missing teams into team column

```python
nl['team'] = nl['club'].where(nl['team'].isnull(), other=nl['team'])
```

Removing unimportant columns

```python
nl2 = nl[['team', 'pos', 'pts', 'year', 'league']].copy()
```

```python
nl2.info()
```

```python
nl2.pos.value_counts()
```

```python
nl2.year.value_counts()
```

```python
nl2.team.value_counts()
```

Removingbrackets (+ contents) from end of names, first checking there are no real team names that will be removed

```python
nl2.team.str.split('(').str[1].unique()
```

```python
nl2.team.str.split('[').str[1].unique()
```

```python
iow = nl2.team.str.contains('\(IOW\)')

nl2[iow]
```

Removing brackets and content but avoiding Newport (IOW)

```python
nl2.loc[~iow, 'team'] = nl2.loc[~iow, 'team'].str.replace('\(.*', '') 
nl2.loc[~iow, 'team'] = nl2.loc[~iow, 'team'].str.replace('\[.*', '') 
```

```python
nl2.team.str.split('(').str[1].unique()
```

```python
nl2.team.str.split('[').str[1].unique()
```

Removing repeated team names e.g. Arsenal ! Arsenal is replaced with Arsenal

```python
nl2.team = nl2.team.str.split('!').str[0]
```

```python
nl2.team.value_counts()
```

Regularising team names (stripping blank space and converting to UPPER case)

```python
nl2['team'].replace('\xc2\xa0', '', inplace = True, regex = True) #removing 'non-breaking' spaces
nl2['team'] = nl2['team'].str.upper().str.strip() #converting to UPPER and stripping spaces
```

```python
nl2.team.value_counts()
```
