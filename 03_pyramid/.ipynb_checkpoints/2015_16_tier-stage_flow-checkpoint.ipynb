Need to regularise tier values

```python
fac_combined.tier.value_counts()
```

```python
def tier_reg(tier):
    if tier > 500:
        tier_new = tier / 100
    else:
        tier_new = tier / 10
    return int(tier_new)

fac_combined['tier'] = fac_combined.tier.apply(tier_reg)
```
